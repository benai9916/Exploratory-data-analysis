## 1. Missing Values

### a. Categorical missing value 

```python
# list out all the NaN values of categorical variable

categorical_column = [cols for cols in house_df.columns if house_df[cols].isnull().sum()> 1 and house_df[cols].dtype == 'O']
categorical_column
```

```python
# print the number of missing values

for cols in categorical_column:
    missing_val = house_df[cols].isnull().sum()
    print('{} has {} missing value'.format(cols, missing_val))
```

```python
# replacing missing value with the most common word

for cols in categorical_column:
    common_category = dict(house_df[cols].value_counts())

    house_df[cols] = house_df[cols].fillna('Missing')
    print(cols ,house_df[cols].isna().sum())
```

```python
# confirm all the missing value are fixedl
list(house_df.isnull().sum())
```

### b. Numerical missing value 

```python
# Now lets check for numerical variables the contains missing values 

numerical_column = [cols for cols in house_df.columns if house_df[cols].isnull().sum()> 1 and house_df[cols].dtype != 'O' and 'Yr' not in cols and 'Year' not in cols]
```

```python
# print the number of missing values

for cols in numerical_column:
    print('{} has {} missing values'.format(cols, house_df[cols].isnull().sum()))
```

**We will replace the missing value with the median insted of mean since there are outlies in the data**

```python
# replacing the missing value with median

for cols in numerical_column:
    house_df[cols] = house_df.fillna(house_df[cols].median())
    
for cols in numerical_column:
    print('{} has {} missing values'.format(cols, house_df[cols].isnull().sum()))
```

### 2. Temporal Variables (Date Time Variables)

```python
# substract the yearsold with other temporal variable

for cols in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    house_df[cols] =  house_df['YrSold']-house_df[cols]
```

```python
# check the columns

house_df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
# fixed skewed data into log normal distribution

skewed_cols = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for cols in skewed_cols:
    house_df[cols] = house_df[cols].astype(int)
    if list(house_df[cols].values) == 0:
        print('There is zero value')
    else:
        house_df[cols] = np.log(house_df[cols].values)
```

```python
house_df[skewed_cols].head()
```

## 3.Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations



```python
categorical_col = [cols for cols in house_df.columns if house_df[cols].dtype == 'O']

categorical_col
```

```python
for cols in categorical_col:
    temp = house_df.groupby(cols)['SalePrice'].count() / len(house_df)
    temp_cols = temp[temp > 0.01].index
    house_df[cols] = np.where(house_df[cols].isin(temp_cols), house_df[cols], 'Rare_value') 
#     print(a.get_group(house_df[cols]))
```

```python
house_df['Electrical'].dropna(inplace=True)
house_df['LotFrontage'].isna().sum()
```

#### Encoding

```python
# covert the categorical varaible to numerical varaible
for feature in categorical_col:
    labels_ordered=house_df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    house_df[feature]=house_df[feature].map(labels_ordered)
```
