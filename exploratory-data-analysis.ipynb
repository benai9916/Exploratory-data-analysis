## Missing values

```python
# here we will check the numbers of missing values in each columns and print those colum which has missing values

for cols in house_df.columns.tolist():
    missing_value = house_df[cols].isnull().sum(axis = 0).tolist()
    
    if missing_value != 0:
        print(cols , 'has' ,  missing_value , 'Missing value')
```

### Since they are many missing values, we need to find the relationship between missing values and Sales Price

```python
# will find the relationship between out target varaible(sale price) and missing vaue (0) and non-missing value(1)
missing_value = [feature for feature in house_df.columns if house_df[feature].isnull().sum()]

for mis_val_col in missing_value:
    house_df_copy = house_df.copy(deep=True)
    
    # convert all the missing value to 0 or else 1
    house_df_copy[mis_val_col] = np.where(house_df_copy[mis_val_col].isnull(), 1,0)
    
    #plot the sale price against missing and non-missing values
    house_df_copy.groupby(mis_val_col)['SalePrice'].median().plot.bar()
    plt.show()
```

There is a relationship brtween the missing values and the dependent variable. Whereever there is missing value sale price is high
So We need to replace these missing values with something meaningful. Will cover this in Feature engineering section.

## Numerical values

```python
# dataframe of all columns with numerical values
numeric_col = [num_col for num_col in house_df.columns if house_df[num_col].dtype != 'O']

house_df[numeric_col].head()
```

```python
# list of all those columns which contain years or data (Temporal variables)
year_cols = [y for y in house_df.columns if 'Yr' in y or 'Year' in y]

house_df[year_cols].head(4)
```

```python
# check the data type of the year columns
house_df[year_cols].dtypes
```

**As you can see the year is not in datetime formate we need to convert it into data time formate**

```python
# check the values in  year columns

for val in year_cols:
    print(val, house_df[val].unique(), '\n')
```

```python
# Lets analyze the Datetime columns
# We will check whether there is a relation between year the house is sold and the sales price

for cols in year_cols:
    house_df.groupby(cols)['SalePrice'].median().plot()
    plt.xlabel('Year Sold')
    plt.ylabel('Median House Price')
    plt.title("House Price vs YearSold")
    plt.show()
```

**House price is decreasing in the 'Year Sold' column, will further check why is this happening**

```python
# we will check all difference between year sold and the sale price

for cols in year_cols:
    if cols !=  'YrSold':
        house_df_copy = house_df.copy(deep=True)
        
        # difference between the years
        house_df_copy[cols] = house_df_copy['YrSold'] - house_df_copy[cols]
        
        sns.scatterplot(house_df_copy[cols], house_df['SalePrice'])
        plt.show()
```

**We can visualize that as the house get older the sale price drop**

```python
# Numerical variables are usually of 2 type
# 1. Continous variable and Discrete Variables

discrete_cols = [cols for cols in numeric_col if len(house_df[cols].unique()) < 25 and cols not in year_cols]

print('The number of discrete valye is:', len(discrete_cols))
```

```python
house_df[discrete_cols].head()
```

```python
# find out the relationship between discrete columns and sales price

for cols in discrete_cols:
    house_df_copy = house_df.copy(deep=True)
    
    house_df_copy[cols] = house_df_copy.groupby(cols)['SalePrice'].median().plot.bar()
    plt.show()
```

**There is a relationship between variable number and SalePrice**

## Continuous Variable

```python
continouse_cols = [cols for cols in numeric_col if cols not in discrete_cols + year_cols]

print('The number of numeric columns are:', len(continouse_cols))
```

```python
# Lets analyse the continuous values by creating histograms to understand the distribution

for cols in continouse_cols:
    plt.hist(house_df[cols], bins=25)
    plt.xlabel(cols)
    plt.ylabel('Counts')
    plt.show()
```

```python
# Since the continouse data does not follow normal distribution we have to do feature scaling

for cols in continouse_cols:
    house_df_copy = house_df.copy(deep=True)
    
    if 0 in house_df_copy[cols].unique():
        pass
    else:
        house_df_copy[cols] = np.log(house_df_copy[cols])
        house_df_copy['SalePrice'] = np.log(house_df_copy['SalePrice'])
        
        plt.scatter(house_df_copy[cols], house_df_copy['SalePrice'])
        plt.xlabel(cols)
        plt.ylabel('Sale price')
        plt.show()
```
